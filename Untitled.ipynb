```python
r_dict = {}
failed_links = []
chrome_options = Options()
chrome_options.add_argument("enable-automation");
chrome_options.add_argument("--headless");
chrome_options.add_argument("--window-size=1920,1080");
chrome_options.add_argument("--no-sandbox");
chrome_options.add_argument("--disable-extensions");
chrome_options.add_argument("--dns-prefetch-disable");
chrome_options.add_argument("--disable-gpu");
chrome_options.add_argument("--headless")
browser  = webdriver.Chrome(ChromeDriverManager().install(),options=chrome_options)
for link in alist:
    name = link.split('/')[-2]
    try:
        browser.get(link)
        ingre_l = browser.find_element(By.CLASS_NAME, 'recipe-ingredients__list')
        direc_l = browser.find_element(By.CLASS_NAME, 'recipe-directions')
        r_dict[name] = {
            'link' : link,
            'ingredients' : [x.get_attribute('innerHTML') for x in ingre_l.find_elements(By.TAG_NAME, 'li')],
            'directions'  : [x.get_attribute('innerHTML') for x in direc_l.find_elements(By.TAG_NAME, 'span')]
        }
        print(r_dict[name])
    except:
        print('Failed on' + link)
        failed_links.append(link)
```

```python
mydb = client['RecipeScrapeSite']
mycollection = mydb['RecipeList']
for key in r_dict.keys():
    ing_l = []
    for ing in r_dict[key]['ingredients']:
        ing_l.append(ing.split('<div')[0])
    mycollection.insert_one({
        'link' : r_dict[key]['link'],
        'ingredients' : ing_l,
        'directions' : r_dict[key]['directions']
    })
```
